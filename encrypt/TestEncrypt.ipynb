#### Test on Taylor series expansion to reproduce exponential

```python
np.math.factorial(0)
```

```python
anum = en1.encrypt(1)
```

```python
def exp(x, n=4):
    d = 1.0
    rt = 0.0
    for i in range(n):
        rt += d/np.math.factorial(i)
        d = x*d
    return rt
```

```python
print("Sample exp 1: ", np.exp(1))
for i in range(1, 10):
    print("n =", i, ":", exp(1, i))
```

```python
ennum = exp(anum, 4)
en1.decrypt(ennum)[0]
```

```python
def HEsoftmax(a):
    return exp(a)/np.sum(exp(a))
```
